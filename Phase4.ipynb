```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
minimum = dataset['age'].min()
print(f"Minimum: {minimum}")
```

```python
maximum = dataset['age'].max()
print(f"Maximum: {maximum}")
```

```python
mean_value = np.mean(dataset['age'])
print(f"Mean: {mean_value}")
```

```python
median_value = np.median(dataset['age'])
print(f"Median: {median_value}")
```

```python
age_values = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]


plt.figure(figsize=(10, 6))
plt.bar(range(len(age_values)), age_values, tick_label=age_values)
plt.title('Bar Graph')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, explode=explode)
plt.title('Pie Chart')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(dataset['age'], bins=20, kde=True)
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Label', y='likes/post', data=dataset)
plt.title('Box PLot')
plt.xlabel('Label')
plt.ylabel('Likes/Post')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Label', data=dataset, palette="Set3")
plt.title('Count Plot')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
stacked_data = dataset.groupby('Label')[['#friends', '#following', '#postshared']].mean()

stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar')
plt.xlabel('Label')
plt.ylabel('Average Values')
plt.legend(title='Features', loc='upper right')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sin(x)')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cos(x)')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tan(x)')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exp(x)')

fig.suptitle('Grid Plot')

plt.tight_layout()

plt.show()
```

```python
correlation_matrix = dataset.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='#friends', y='#following', data=dataset)
plt.title('Scatter Plot')
plt.xlabel('#Friends')
plt.ylabel('#Following')
plt.show()
```
